# Random Forest

```python
dia1.head()
```

```python
i=dia1.iloc[:,0:8].values
d=dia1.iloc[:,2].values
```

```python
from sklearn.model_selection import train_test_split

i_train, i_test, d_train, d_test = train_test_split(i, d, test_size=0.2, random_state=0)
```

```python
from sklearn. preprocessing import StandardScaler

sc = StandardScaler()
i_train = sc.fit_transform(i_train)
i_test = sc.transform(i_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=20, random_state=0)

regressor.fit(i_train, d_train)
d_pred = regressor.predict(i_test)
```

```python
from sklearn.metrics import accuracy_score

acuRF=accuracy_score(d_test, d_pred)
print(acuRF)
print ("\nAccuracy: " + "{:.2%}".format(acuRF))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(d_test,d_pred)
class_names = [0,1]
fig,ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks,class_names)
plt.yticks(tick_marks,class_names)
sns.heatmap(pd.DataFrame(confusion_matrix), annot = True, cmap = 'Greens', fmt = 'g')
ax.xaxis.set_label_position('top')
plt.tight_layout()
plt.title('Confusion matrix for Random Forest')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```
